```python
#Import relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame()

# Access the data, csv file
df = pd.read_csv('Spotify_dataset.csv')
#df = pd.read_csv('https://raw.githubusercontent.com/rafaelf89/PFDAFinalProject/main/Spotify_dataset.csv')
```

```python
#Import and select relevant columns for my data.

spotifydata = pd.read_csv('Spotify_dataset.csv')
spotifydata=spotifydata.drop(['artists','popularity','album_name','track_name','track_id'], axis = 1)
```

```python
### Part2 Calculate mean, median, and mode

mode_art = df['artists'].mode()
print("The mode of artists is: " , mode_art[0])
print()

### Mode, median, and mode for artist
mean_pop = df['popularity'].mean()
median_pop = df['popularity'].median()
mode_pop = df['popularity'].mode()

print("The mean of popularity is: " , mean_pop)
print("The median of popularity is: " , median_pop)
print("The mode of popularity is: " , mode_pop[0])
print()

### Mode for album_name
mode_album = df['album_name'].mode()
print("The mode of album_name is: " , mode_album[0])
print()

### Mode for track_name
mode_track = df['track_name'].mode()
print("The mode of track_name is: " , mode_track[0])
print()
```

```python
dfpopularity = df[df['popularity'] > 75]
dfpopularity = dfpopularity[dfpopularity['popularity'] < 100]

dfpopularity
```

```python
#Graphic to show the Top 15 Artists

df['artists'].value_counts().nlargest(15).plot(kind='bar', title = 'Top 15 Artists', color='orange')
plt.xlabel('Top 15 Artists')
```

```python
#Graphic to show the Top 15 Albums

df['album_name'].value_counts().nlargest(15).plot(kind='bar', title = 'Top 15 Albums', color = 'c')
plt.xlabel('Top 15 Album')
```

```python
#Graphic to show the Top 15 Tracks

df['track_name'].value_counts().nlargest(15).plot(kind='bar', title = 'Top 15 Tracks', color = 'brown')
plt.xlabel('Top 15 Tracks')
```
